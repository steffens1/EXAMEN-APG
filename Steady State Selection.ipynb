```python
def dec_bin(signo , entero, decimal, print_):
    cadena = ""
    cadena =  cadena + str(signo)
    bin_entero = bin(entero)
    cadena =  cadena + (2 - len(bin_entero[2:]))*"0" + str(bin_entero[2:])
    bin_decimal= bin(decimal)
    cadena =  cadena + ( 10 - len(bin_decimal[2:12]))*"0" + str(bin_decimal[2:12])
    if print_ : 
        print(cadena)
    return cadena
```

```python
def bin_decimal (x, print_):
    signo =  int(x[0])
    
    entero = ''.join(str(x[e]) for e in range(1, 2))
    decimal= ''.join(str(x[e]) for e in range(2, len(x)))
    
    entero = int(entero,2)
    decimal= int(decimal,2)
    decimal= float (decimal / 1000)    
    resul = str(entero)
    if signo==0:
        if print_: print(float(resul) + decimal)
        return float(resul) + decimal
    else:
        if print_: print(-float(resul) - decimal)
        return -float(resul) - decimal
```

```python
def distribuir(x):    
    if x <=0 : 
        signo = 1
        cadena = str(x)
        entero = int(cadena[1])
        decimal = int(cadena [3:6])
    else: 
        signo = 0
        cadena = str (x)
        entero = int(cadena[0])
        decimal = int(cadena [2:5])
    
    return signo , entero , decimal 
```

```python
def funcion(x1 , x2 ):
    if  x1 >= -2.048 and x1 <= 2.048:
        x1 = x1
    else :
        x1 = 0
    if  x2 >= -2.048 and x2 <= 2.048:
        x2=x2
    else :
        x2=0
    return 100*(float(x1**2) - float(x2**2))  + float((1 - x1)**2)
```

```python
def unificar( x1 , x2 ):
    signo1 , entero1 , decimal1 = distribuir(x1)
    signo2 , entero2 , decimal2 = distribuir(x2)
    
    cromosoma1 = dec_bin(signo1, entero1 , decimal1 , False)
    cromosoma2 = dec_bin(signo2, entero2 , decimal2 , False)
    
    resul = { 
        "c1" : cromosoma1 ,
        "c2" : cromosoma2
    }
    return resul
```

```python
def list_string(lista):
    cadena=""
    for i in lista:
        cadena = cadena + str(i)
    return cadena
```

```python
class Individuo(object):
    def __init__(self, cromosoma):
        self.cromosoma = cromosoma  
        self.fitness = self.cal_fitness()
        
    @classmethod
    def mutar_gen( self ): 
        gene = random.randrange(0,1)
        return gene 
    
    @classmethod
    def crear_gnoma( self ): 
        x1= random.uniform(-2.48 , 2.48 )
        x2= random.uniform(-2.48 , 2.48 )
        cromosoma = unificar(x1 , x2 )
        return cromosoma
    
    def cruzar(self, par2): 
        #cruce por proabilidad
        hijo1_cromosoma = [] 
        for gp1, gp2  in zip(self.cromosoma["c1"], par2.cromosoma["c1"]):
            probabilidad = random.random()
            if probabilidad < 0.45 :
                hijo1_cromosoma.append(gp1)
            elif probabilidad < 0.90 :
                hijo1_cromosoma.append(gp2)
            else :
                hijo1_cromosoma.append(self.mutar_gen())
        
        hijo2_cromosoma = [] 
        for gp1, gp2  in zip(self.cromosoma["c2"], par2.cromosoma["c2"]):
            probabilidad = random.random()
            if probabilidad < 0.45 :
                hijo2_cromosoma.append(gp1)
            elif probabilidad < 0.90 :
                hijo2_cromosoma.append(gp2)
            else :
                hijo2_cromosoma.append(self.mutar_gen())        
        
        resul = { "c1" : list_string(hijo1_cromosoma) ,
            "c2" : list_string(hijo2_cromosoma) } 
        
        return Individuo(resul)
    
    def cal_fitness(self): 
        x1 = bin_decimal(self.cromosoma["c1"] , False)
        x2 = bin_decimal(self.cromosoma["c2"] , False)
        fitness =  funcion(x1 , x2)
        return fitness
```
