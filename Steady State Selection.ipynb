```python
def dec_bin(signo , entero, decimal, print_):
    cadena = ""
    cadena =  cadena + str(signo)
    bin_entero = bin(entero)
    cadena =  cadena + str(bin_entero[2:])
    bin_decimal= bin(decimal)
    cadena =  cadena + str(bin_decimal[2:8])
    
    if print_ : 
        print(cadena)
    
    return cadena
```

```python
def bin_decimal (x, print_):
    signo =  x[0]
    entero = ''.join(str(x[e]) for e in range(1, 3))
    decimal= ''.join(str(x[e]) for e in range(3, len(x)))
    
    entero = int(entero,2)
    decimal= int(decimal,2)
    
    if print_:
        print(str(int(entero))  +"." + str(decimal))
        
    resul = str(entero) + "."  + str(decimal)
    if signo == "0":
        return float(resul)
    else:
        return -float(resul)
```

```python
def distribuir(x):    
    if x <=0 : 
        signo = 1
        cadena = str(x)
        entero = int(cadena[1])
        decimal = int(cadena [3:])
    else: 
        signo = 0
        cadena = str (x)
        entero = int(cadena[0])
        decimal = int(cadena [2:])
    
    return signo , entero , decimal 
```

```python
def funcion(x1 , x2 ):
    return 100*(float(x1**2) - float(x2**2))  + float((1 - x1)**2)
```

```python
def unificar( x1 , x2 ):
    signo1 , entero1 , decimal1 = distribuir(x1)
    signo2 , entero2 , decimal2 = distribuir(x2)
    
    cromosoma1 = dec_bin(signo1, entero1 , decimal1 , False)
    cromosoma2 = dec_bin(signo2, entero2 , decimal2 , False)
    
    resul = { 
        "c1" : cromosoma1 ,
        "c2" : cromosoma2
    }
    return resul
```

```python

class Individuo(object):
    
    def __init__(self ,cromosoma): 
        self.cromosoma = cromosoma  
        self.fitness = self.cal_fitness()
        
    @classmethod
    def mutar_gen( self ): 
        gene = random.randrange(0,1)
        return gene 
    
    @classmethod
    def crear_gnoma( self ): 
        x1= random.uniform(-2.48 , 2.48 )
        x2= random.uniform(-2.48 , 2.48 )
        cromosoma = unificar(x1 , x2 )
        return cromosoma
    
    def cruzar(self, par2): 
        #cruce por proabilidad
        hijo1_cromosoma = [] 
        for gp1, gp2  in zip(self.cromosoma["c1"], par2.cromosoma["c1"]):
            probabilidad = random.random()
            if probabilidad < 0.45 :
                hijo1_cromosoma.append(gp1)
            elif probabilidad < 0.90 :
                hijo1_cromosoma.append(gp2)
            else :
                hijo1_cromosoma.append(self.mutar_gen())
        
        hijo2_cromosoma = [] 
        for gp1, gp2  in zip(self.cromosoma["c2"], par2.cromosoma["c2"]):
            probabilidad = random.random()
            if probabilidad < 0.45 :
                hijo1_cromosoma.append(gp1)
            elif probabilidad < 0.90 :
                hijo1_cromosoma.append(gp2)
            else :
                hijo1_cromosoma.append(self.mutar_gen())
                
        resul = { 
            "c1" : hijo1_cromosoma ,
            "c2" : hijo2_cromosoma
        }        
        
        return Individuo(resul)
    
    def cal_fitness(self): 
        print((self.cromosoma["c1"])
        x1 =  bin_decimal(self.cromosoma["c1"] , False)
        x2 =  bin_decimal(self.cromosoma["c2"] , False)
        fitness =  funcion(x1 , x2)
        return fitness
```

```python
def main(): 
    global TAM_POBLACION  
  
    generacion = 1
  
    encontrado = False
    poblacion = [] 
     
    for _ in range(TAM_POBLACION ): 
                gnome = Individuo.crear_gnoma() 
                poblacion.append(Individuo(gnome))
                
    for i in poblacion:
        print(len(i.cromosoma["c2"]))
        print(bin_decimal(i.cromosoma["c2"], False))
                
    for _ in range(1, 1): 
        #se ordena la poblacion segun atributo fitness
        poblacion = sorted(poblacion, key = lambda x:x.fitness) 
            
        nueva_generacion = [] 
        
        for _ in range(int(TAM_POBLACION/2)): 
            
            #padres seleccionados aleatoriamente del 50% mejor clasificados la generacion actual 
            
            padre1 = random.choice(poblacion[:int(TAM_POBLACION/2)]) 
            padre2 = random.choice(poblacion[:int(TAM_POBLACION/2)]) 
            
            #se cruza los padres 
            (hijo) = padre1.cruzar(padre2)
            
            #stady state selection se escoge de los 2 padres y el hijo, segun el atributo fitness , se da la opcion de que los padres a la siguiente generacion
            lista = [ hijo  , padre1 , padre2]
            lista = sorted(lista , key = lambda x:x.fitness)
            
            #se introduce a la nueva generacion al mejor
            nueva_generacion.append(lista[0])
        
        poblacion = nueva_generacion 
        print("generacion: {} \tString: {} \tFitness: {} ".format(generacion,"".join(poblacion[0].cromosoma),poblacion[0].fitness)) 
        generacion += 1
        
    print("generacion: {}\tString: {}\tFitness: {}".format(generacion,"".join(poblacion[0].cromosoma), poblacion[0].fitness)) 
```
